```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(x):
    return np.maximum(0,x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-5,5,100)
y = relu(x)

plt.figure(figsize=(8,4))
plt.plot(x, y, label='ReLU(x)', color='blue')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.legend()
plt.grid(True)
#plt.savefig("relu_plot_dpi300.png", dpi=300)
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = sigmoid(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Sigmoid(x)', color='green')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.legend()
plt.grid(True)
plt.yticks([0, 0.25, 0.5, 0.75, 1])
plt.savefig('sigmoid_plot_dpi300.png', dpi=300)

plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = tanh(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='tanh(x)', color='orange')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.legend()
plt.grid(True)

plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.savefig('tanh_plot_dpi300.png', dpi=300)
plt.show()
```

```python
import random

G = nx.DiGraph()

nodes = ['I_1', 'I_2', 'I_3', 'I_4'] + \
        ['H_1-' + str(i) for i in range(1, 7)] + \
        ['H_2-' + str(i) for i in range(1, 7)] + \
        ['O_1', 'O_2', 'O_3']
G.add_nodes_from(nodes)

edges = [('I_' + str(i), 'H_1-' + str(j)) for i in range(1, 5) for j in range(1, 7)] + \
        [('H_1-' + str(i), 'H_2-' + str(j)) for i in range(1, 7) for j in range(1, 7)] + \
        [('H_2-' + str(i), 'O_' + str(j)) for i in range(1, 7) for j in range(1, 4)]
G.add_edges_from(edges)

nodes_to_drop_h1 = random.sample(['H_1-1', 'H_1-3', 'H_1-5'], 2)
nodes_to_drop_h2 = random.sample(['H_2-2', 'H_2-4', 'H_2-6'], 2)
G.remove_nodes_from(nodes_to_drop_h1 + nodes_to_drop_h2)

pos = {
    'I_4': (0, 15),
    'I_3': (0, 12),
    'I_2': (0, 9),
    'I_1': (0, 6),
}
pos.update({f'H_1-{i}': (1, i * 3) for i in range(1, 7)})
pos.update({f'H_2-{i}': (2, i * 3) for i in range(1, 7)})
pos.update({f'O_{i}': (3, i*3 + 4.5) for i in range(1, 4)})

nx.draw(G, pos, with_labels=True, node_size=1000, node_color='lightblue', font_size=10, font_weight='bold')

for node in nodes_to_drop_h1 + nodes_to_drop_h2:
    plt.text(pos[node][0], pos[node][1], 'X', fontsize=12, color='red', fontweight='bold', ha='center', va='center')

#plt.savefig('mlp_dropout_example.png', dpi=300)
plt.show()
```

```python
import tensorflow as tf

tf.random.set_seed(0)

X = tf.random.uniform(minval=-3, maxval=3, shape=(200,))
y = 3 * tf.math.sin(X) + tf.random.normal(mean=0, stddev=0.5, shape=(200,))
```

```python
plt.figure(figsize=(10, 10))

plt.scatter(X, y, label='Data points', s=10)
#plt.savefig('before_fit.png', dpi=300)
plt.show()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(loss=tf.keras.losses.mean_squared_error)

model.fit(
    tf.reshape(X, shape=(200, 1)),
    tf.reshape(y, shape=(200, 1)),
    batch_size=32,
    epochs=20
)
```

```python
predictions = model(tf.reshape(X, shape=(200, 1)))

X_np = np.array(X)
mean_np = np.array(predictions)

sorted_indices = np.argsort(X_np)
sorted_X = X_np[sorted_indices]
sorted_mean = mean_np[sorted_indices]

plt.figure(figsize=(10, 10))
dot_size = 2

plt.scatter(X_np, y, label='Data points', color='blue', s=2)
plt.plot(sorted_X, sorted_mean, color='red')
print(sorted_mean.shape)
#plt.savefig('overfit.png', dpi=300)
plt.show()
```

```python
from PIL import Image

def parse_kitti_label(label_str):
    parts = label_str.split()
    label_type = parts[0]
    box2d = [float(parts[i]) for i in range(4, 8)]
    return label_type, box2d

def draw_labeled_image(image, labels, include_dontcare=False, save_fig_path=None, dpi=300):
    fig, ax = plt.subplots(figsize=(20, 20)) 
    ax.imshow(image)
    
    label_types = set()
    for label in labels:
        label_type, _ = parse_kitti_label(label)
        if not include_dontcare and label_type == 'DontCare':
            continue
        label_types.add(label_type)
    
    num_labels = len(label_types)
    color_map = plt.get_cmap('tab20', num_labels)
    
    label_colors = {
        'Car': 'red',
        'Pedestrian': 'lightblue',
        'Truck': 'teal',
        'Van': 'blue',
        'DontCare': 'black',
        'Cyclist': 'purple',
        'Tram': 'green',
        'Misc': 'magenta'
    }
    
    for label in labels:
        label_type, box2d = parse_kitti_label(label)
        
        if not include_dontcare and label_type == 'DontCare':
            continue
        
        color = label_colors.get(label_type, 'k')  
        rect = patches.Rectangle(
            (box2d[0], box2d[1]),
            box2d[2] - box2d[0],
            box2d[3] - box2d[1],
            linewidth=1,
            edgecolor=color,
            facecolor='none'
        )
        ax.add_patch(rect)
        
        ax.text(
            box2d[0],
            box2d[1],
            f'{label_type}',
            fontsize=12,
            color="black",
            verticalalignment='bottom',
            bbox={'facecolor': "teal", 'alpha': 1, 'pad': 2}
        )
    
    ax.set_xticks([])
    ax.set_yticks([])
    
    if save_fig_path:
        plt.savefig(save_fig_path, dpi=dpi, bbox_inches='tight')
    else:
        plt.show()
    plt.close()

image_path = 'data_object_image_2/training/image_2/000177.png'
labels_path = 'data_object_label_2/training/label_2/000177.txt'

with open(labels_path, 'r') as file:
    labels = file.readlines()

image = Image.open(image_path)
draw_labeled_image(
    image,
    labels,
    include_dontcare=True,
    #save_fig_path="example_kitti_image_3.png",
    dpi=300
)
```
